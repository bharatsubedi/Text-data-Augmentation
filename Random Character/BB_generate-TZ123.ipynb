```python
def image_generate():
    img_number = 1
    imagefiles = [file for file in os.listdir("PPMNoBright/") if file.endswith('.ppm')]
    for i in range(10000):
        liss = np.random.choice(imagefiles,1,replace= False)
        lis =liss[0]
        name = "Nobri_New_"+str(img_number)
        fontsize = [240,239,240,238,240]
        
        col_1 = np.random.randint(50,55)
        col_2 = np.random.randint(53,55)
        col_3 = np.random.randint(53,55)
        color = (col_1,col_2,col_3)
        LW = np.random.randint(100,120)
        LH = np.random.randint(200,700)
        gap1 =np.random.randint(90,110)
        gap2 =np.random.randint(90,110)
        
        # gap bet char
        char_gap_1=np.random.randint(0,20)
        char_gap_2=np.random.randint(0,20)
        char_gap_3=np.random.randint(-2,10)
        char_gap_4=np.random.randint(0,20)
        
        updown = [0,0,0]
        up = np.random.choice(updown,2,replace= False)
        upvalue1 =up[0]
        upvalue2 =up[1]
        down = np.random.choice(updown,2,replace= False)
        downvalue1 =down[0]
        downvalue2 =down[1]
        
        #image =Image.new('RGBA', (1000, 300), (255,255,255))
        image = Image.open("PPMNoBright/{}".format(lis))
        # image shifting 
        a,b = 1,0
        c = [0,1] #left/right (i.e. 5/-5)
        LR_select = np.random.choice(c,1,replace = False)
        LR = LR_select[0]
        d,e = 0,1
        f = [0,1] #up/down (i.e. 5/-5)
        UD_select = np.random.choice(c,1,replace = False)
        UD = UD_select[0]
        image = image.transform(image.size, Image.AFFINE, (a, b, LR, d, e, UD))
        
        new_gt="gt_"
        img_width, img_height = image.size   # Get dimensions
        with open('Art_Dataset/Nobri_New_gt/{}{}.txt'.format(new_gt,name),'w',encoding='utf-8') as txtfile:
            drawer = ImageDraw.Draw(image)

            flists =[f for f in listdir("OCR-Fonts/") if isfile(join("OCR-Fonts/", f))]
              # dynamic choice of font 
            fontlist = np.random.choice(flists,1,replace = False)
            fsize = np.random.choice(fontsize,2,replace = False)
            image_size = (img_width,img_height)
            #print(fontlist[0])
            #print(fsize[0])
            
            font = ImageFont.truetype("OCR-Fonts/"+fontlist[0], fsize[0])
            ascent, descent = font.getmetrics()
           # print("ascent::", ascent)
            #print("descent::", descent)
            string ="ABCDEFGHIJKLMNOPQ12345RSTUVWXYZ"
            string_list = list(string)
            string_first =  np.random.choice(string_list,2,replace = False)
            string_second =  np.random.choice(string_list,2,replace = False)
            string_third =  np.random.choice(string_list,3,replace = False)
            
            first = string_first[0]
            second = string_first[1]
            third = string_second[0]
            fourth = string_second[1]
            fifth = string_third[0]
            sixth = string_third[1]
            seventh = string_third[2]
            
            ###  first word 
            width1, height1 = drawer.textsize( first, font=font,direction="ltr")
            (width, baseline), (offset_x, offset_y) = font.font.getsize(first)
           # print("width::", width)
            #print("baseline::", baseline)
            #print("offset_x::", offset_x)
            #print("offset_y::", offset_y)
            ab = font.getmask(first).getbbox()
            #print("ab::", ab)
            #print("width1::", width1)
            #print("height1::", height1)
            offset = font.getoffset(first)
           # print("offset10::", offset[0])
            #print("offset11::", offset[1])
            left_width1, left_height1 =LW,LH   
            drawer.text(( left_width1,  left_height1), first, fill=color, font=font,direction="ltr")
            # drawing rectangle surrounding text
           
            lengthx1 = left_width1+ab[2] #+ width1 
            lengthy1 = left_height1 + height1
            
           
            txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width1+ab[0],left_height1+offset[1],
                                                                lengthx1,left_height1+offset[1],
                                                                lengthx1,lengthy1,
                                                                left_width1+ab[0],lengthy1,first
                                                               ))
            #print('lengthx1:{}'.format(lengthx1))
            ##### second word
            width2, height2 = drawer.textsize( second, font=font,direction="ltr")
            ab = font.getmask(second).getbbox()
            offset = font.getoffset(second)
           # print("offset20::", offset[0])
            #print("offset21::", offset[1])
            left_width2, left_height2 =lengthx1-ab[0]+char_gap_1,left_height1
            drawer.text(( left_width2,  left_height2), second, fill=color, font=font,direction="ltr")
            # drawing rectangle surrounding text
           
            lengthx2 = left_width2 +ab[2]
            lengthy2 = left_height2 + height2
            #print('lengthx2:{}'.format(lengthx2))
            txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width2+ab[0],left_height2+offset[1],
                                                                lengthx2,left_height2+offset[1],
                                                                lengthx2,lengthy2,
                                                                left_width2+ab[0],lengthy2,second
                                                               ))
            #print('lengthx2:{}'.format(lengthx2))
            ######  third word
            width3, height3 = drawer.textsize( third, font=font,direction="ltr")
            ab = font.getmask(third).getbbox()
            left_width3, left_height3 =lengthx2+gap1-ab[0],left_height2  
            drawer.text(( left_width3,  left_height3), third, fill=color, font=font,direction="ltr")
            # drawing rectangle surrounding text
            offset = font.getoffset(third)
            lengthx3 = left_width3 +ab[2]
            lengthy3 = left_height3 + height3
            txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width3+ab[0],left_height3+offset[1],
                                                                lengthx3,left_height3+offset[1],
                                                                lengthx3,lengthy3,
                                                                left_width3+ab[0],lengthy3,third
                                                               ))
            
            ######  fourth word
            width4, height4 = drawer.textsize( fourth, font=font,direction="ltr")
            ab = font.getmask(fourth).getbbox()
            left_width4, left_height4 =lengthx3-ab[0]+char_gap_2,left_height3 
            drawer.text(( left_width4,  left_height4), fourth, fill=color, font=font,direction="ltr")
            # drawing rectangle surrounding text
            offset = font.getoffset(fourth)
            lengthx4 = left_width4 + ab[2]#+1
            lengthy4 = left_height4 + height4
            txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width4+ab[0],left_height4+offset[1],
                                                                lengthx4,left_height4+offset[1],
                                                                lengthx4,lengthy4,
                                                                left_width4+ab[0],lengthy4,fourth
                                                               ))
            
            ######  fifth word
            width5, height5 = drawer.textsize( fifth, font=font,direction="ltr")
            ab = font.getmask(fifth).getbbox()
            offset = font.getoffset(fifth)
            offset6 = font.getoffset(sixth)
            #print("width5::", width5)
            left_width5, left_height5 =lengthx4+gap2-ab[0],left_height4 
            drawer.text(( left_width5,  left_height5), fifth, fill=color, font=font,direction="ltr")
            # drawing rectangle surrounding text
            lengthx5 = left_width5 +ab[2]
            lengthy5 = left_height5 + height5
            txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width5+ab[0],left_height5+offset[1],
                                                                lengthx5,left_height5+offset[1],
                                                                lengthx5,lengthy5,
                                                                left_width5+ab[0],lengthy5,fifth
                                                               ))
            
            ######  sixth word
            width6, height6 = drawer.textsize( sixth, font=font,direction="ltr")
            ab = font.getmask(sixth).getbbox()
            offset = font.getoffset(sixth)
            left_width6, left_height6 =lengthx5-ab[0]+char_gap_3,left_height5 
            drawer.text(( left_width6,  left_height6), sixth, fill=color, font=font,direction="ltr")
            # drawing rectangle surrounding text
            
            lengthx6 = left_width6 +ab[2]
            lengthy6 = left_height6 + height6
            txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width6+ab[0],left_height6+offset[1],
                                                                lengthx6,left_height6+offset[1],
                                                                lengthx6,lengthy6,
                                                                left_width6+ab[0],lengthy6,sixth
                                                               ))
            
            ######  seventh word
            width7, height7 = drawer.textsize( seventh, font=font,direction="ltr")
            ab = font.getmask(seventh).getbbox()
            size = font.getsize(seventh)
            left_width7, left_height7 =lengthx6-ab[0]+char_gap_4,left_height6 
            drawer.text(( left_width7,  left_height7), seventh, fill=color, font=font,direction="ltr")
            # drawing rectangle surrounding text
            offset = font.getoffset(seventh)
            lengthx7 = left_width7 +ab[2]
            lengthy7 = left_height7 + height7
            txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width7+ab[0],left_height7+offset[1],
                                                                lengthx7,left_height7+offset[1],
                                                                lengthx7,lengthy7,
                                                                left_width7+ab[0],lengthy7,seventh
                                                               ))
        image.save('Art_Dataset/Nobri_New_img/{}.png'.format(name))
        img_number+=1
```
