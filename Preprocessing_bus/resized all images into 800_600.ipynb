```python
import cv2
import os
import numpy as np
from PIL import Image
```

```python
imagelist = [file for file in os.listdir("Art_Dataset/without_number_gray/") if file.endswith('.png') or file.endswith('.jpg')]
print(len(imagelist))
if not os.path.exists("Art_Dataset/train_data/"):
    os.mkdir("Art_Dataset/train_data/")
```

```python
for image in imagelist:
    print(image)
    img = cv2.imread("Art_Dataset/without_number_gray/{}".format(image))
    img_gray = cv2.resize(img,(800,600))
    new_name = image[:-4]
    cv2.imwrite("Art_Dataset/train_data/{}.png".format(new_name),img_gray)
```

```python
###############  ground truth resize ###########################
```

```python
gtfiles = [file for file in os.listdir("Art_Dataset/without_number_gt/") if file.endswith('.txt')]
print(len(gtfiles))
if not os.path.exists("Art_Dataset/gt/"):
    os.mkdir("Art_Dataset/gt/")
```

```python
for gt in gtfiles:  
    with open("Art_Dataset/gt/{}".format(gt),"w") as f:
        with open("Art_Dataset/without_number_gt/{}".format(gt),"r") as readgt:
            for line in readgt.readlines():
                cordinate = line.split(',')
                x1 = int(cordinate[0])//2
                x2 = int(cordinate[1])//2
                x3 = int(cordinate[2])//2
                x4 = int(cordinate[3])//2
                x5 = int(cordinate[4])//2
                x6 = int(cordinate[5])//2
                x7 = int(cordinate[6])//2
                x8 = int(cordinate[7])//2
                word = str(cordinate[8])
                f.write("{},{},{},{},{},{},{},{},{}".format(x1,x2,x3,x4,x5,x6,x7,x8,word))
```

```python
imagelists = [ file for file in os.listdir("Art_Dataset/train_data/") if file.endswith('.png')]
gtlists = [gt for gt in os.listdir("Art_Dataset/gt/") if gt.endswith('.txt')]
print(len(gtlists))
print(gtlists[0])
print(len(imagelists))
if not os.path.exists("Art_Dataset/crop_img/"):
                os.makedirs("Art_Dataset/crop_img/")

```

```python
number = 1
for image in imagelists:
    img = Image.open("Art_Dataset/train_data/"+image)
    print(image)
    for gt in gtlists:
        new_gt=gt[3:][:-4]
        if (new_gt==image[:-4]):
            f= open("Art_Dataset/gt/"+gt,'r',encoding="utf-8")
            for line in f.read().splitlines():
                coordinate = line.split(',')
                print(coordinate)
                x1 = int(coordinate[0])
                y1 = int(coordinate[1])
                x2 =int(coordinate[4])
                y2 =int(coordinate[5])
                word = coordinate[-1]
                crop_img = img.crop((x1,y1,x2,y2))
                crop_img.save("Art_Dataset/crop_img/{}_{}.jpg".format(new_gt,number),"PNG")  
                number+=1
```

```python
imagelist = [file for file in os.listdir("Art_Dataset/crop_img/") if file.endswith('.png') or file.endswith('.jpg')]
print(len(imagelist))
if not os.path.exists("Art_Dataset/final_crop/"):
    os.mkdir("Art_Dataset/final_crop/")
```

```python
for image in imagelist:
    img = cv2.imread("Art_Dataset/crop_img/{}".format(image))
    new_name = image[:-4]
    cv2.imwrite("Art_Dataset/final_crop/{}.png".format(new_name),img)
```
