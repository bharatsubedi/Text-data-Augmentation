```python
def image_generate():
    img_number = 1
    imagefiles = [file for file in os.listdir("number_img/") if file.endswith('.PNG')]
    for i in range(1000):
        with open('bus_route_number.txt') as f:
            for line in f.read().splitlines():
                liss = np.random.choice(imagefiles,1,replace= False)
                lis =liss[0]
                name = str(img_number)+'_'
                fontsize = [85,80,90,87,82]

                col_1 = np.random.randint(50,55)
                col_2 = np.random.randint(53,55)
                col_3 = np.random.randint(53,55)
                color = (col_1,col_2,col_3)

                # gap bet char
                char_gap_1=np.random.randint(12,15)
                char_gap_2=np.random.randint(12,15)
                char_gap_3=np.random.randint(12,15)
                char_gap_4=np.random.randint(11,15)

                #image =Image.new('RGBA', (1000, 300), (255,255,255))
                image = Image.open("number_img/{}".format(lis))
                if len(line)==2:
                    first = line[0]
                    second = line[1]
                    str_1 =name+first+second
                    new_gt="gt_"
                    img_width, img_height = image.size   # Get dimensions
                    with open('Art_Dataset/Nobri_New_gt/{}{}.txt'.format(new_gt,str_1),'w',encoding='utf-8') as txtfile:
                        drawer = ImageDraw.Draw(image)

                        flists =[f for f in listdir("OCR-Fonts/") if isfile(join("OCR-Fonts/", f))]
                          # dynamic choice of font 
                        fontlist = np.random.choice(flists,1,replace = False)
                        fsize = np.random.choice(fontsize,2,replace = False)
                        image_size = (img_width,img_height)

                        font = ImageFont.truetype("OCR-Fonts/"+fontlist[0], fsize[0])
                        ascent, descent = font.getmetrics()

                        ###  first word 
                        width1, height1 = drawer.textsize( first, font=font)
                        (width, baseline), (offset_x, offset_y) = font.font.getsize(first)
                        ab = font.getmask(first).getbbox()
                        offset = font.getoffset(first)
                        left_width1, left_height1 =20,1   
                        drawer.text(( left_width1,  left_height1), first, fill=color, font=font)
                        # drawing rectangle surrounding text

                        lengthx1 = left_width1+ab[2] #+ width1 
                        lengthy1 = left_height1 + height1


                        txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width1+ab[0],left_height1+offset[1],
                                                                            lengthx1,left_height1+offset[1],
                                                                            lengthx1,lengthy1,
                                                                            left_width1+ab[0],lengthy1,first
                                                                           ))
                        ##### second word
                        width2, height2 = drawer.textsize( second, font=font)
                        ab = font.getmask(second).getbbox()
                        offset = font.getoffset(second)
                        left_width2, left_height2 =lengthx1-ab[0]+char_gap_1,left_height1
                        drawer.text(( left_width2,  left_height2), second, fill=color, font=font)
                        # drawing rectangle surrounding text

                        lengthx2 = left_width2 +ab[2]
                        lengthy2 = left_height2 + height2
                        txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width2+ab[0],left_height2+offset[1],
                                                                            lengthx2,left_height2+offset[1],
                                                                            lengthx2,lengthy2,
                                                                            left_width2+ab[0],lengthy2,second
                                                                           ))
                    
                    image.save('Art_Dataset/Nobri_New_img/{}.png'.format(str_1))
                    img_number+=1
                if len(line)==3:
                    first = line[0]
                    second = line[1]
                    third = line[2]
                    str_2 =name+first+second+third
                    new_gt="gt_"
                    img_width, img_height = image.size   # Get dimensions
                    with open('Art_Dataset/Nobri_New_gt/{}{}.txt'.format(new_gt,str_2),'w',encoding='utf-8') as txtfile:
                        drawer = ImageDraw.Draw(image)

                        flists =[f for f in listdir("OCR-Fonts/") if isfile(join("OCR-Fonts/", f))]
                          # dynamic choice of font 
                        fontlist = np.random.choice(flists,1,replace = False)
                        fsize = np.random.choice(fontsize,2,replace = False)
                        image_size = (img_width,img_height)

                        font = ImageFont.truetype("OCR-Fonts/"+fontlist[0], fsize[0])
                        ascent, descent = font.getmetrics()

                        ###  first word 
                        width1, height1 = drawer.textsize( first, font=font)
                        (width, baseline), (offset_x, offset_y) = font.font.getsize(first)
                        ab = font.getmask(first).getbbox()
                        offset = font.getoffset(first)
                        left_width1, left_height1 =15,1   
                        drawer.text(( left_width1,  left_height1), first, fill=color, font=font)
                        # drawing rectangle surrounding text

                        lengthx1 = left_width1+ab[2] #+ width1 
                        lengthy1 = left_height1 + height1


                        txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width1+ab[0],left_height1+offset[1],
                                                                            lengthx1,left_height1+offset[1],
                                                                            lengthx1,lengthy1,
                                                                            left_width1+ab[0],lengthy1,first
                                                                           ))
                        ##### second word
                        width2, height2 = drawer.textsize( second, font=font)
                        ab = font.getmask(second).getbbox()
                        offset = font.getoffset(second)
                        left_width2, left_height2 =lengthx1-ab[0]+char_gap_1,left_height1
                        drawer.text(( left_width2,  left_height2), second, fill=color, font=font)
                        # drawing rectangle surrounding text

                        lengthx2 = left_width2 +ab[2]
                        lengthy2 = left_height2 + height2
                        txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width2+ab[0],left_height2+offset[1],
                                                                            lengthx2,left_height2+offset[1],
                                                                            lengthx2,lengthy2,
                                                                            left_width2+ab[0],lengthy2,second
                                                                           ))
                        #print('lengthx2:{}'.format(lengthx2))
                        ######  third word
                        width3, height3 = drawer.textsize( third, font=font)
                        ab = font.getmask(third).getbbox()
                        left_width3, left_height3 =lengthx2+char_gap_3-ab[0],left_height2  
                        drawer.text(( left_width3,  left_height3), third, fill=color, font=font)
                        # drawing rectangle surrounding text
                        offset = font.getoffset(third)
                        lengthx3 = left_width3 +ab[2]
                        lengthy3 = left_height3 + height3
                        txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width3+ab[0],left_height3+offset[1],
                                                                            lengthx3,left_height3+offset[1],
                                                                            lengthx3,lengthy3,
                                                                            left_width3+ab[0],lengthy3,third
                                                                           ))

                    image.save('Art_Dataset/Nobri_New_img/{}.png'.format(str_2))
                    img_number+=1
                
                if len(line)==4:
                    first = line[0]
                    second = line[1]
                    third = line[2]
                    fourth = line[3]
                    str_3 =name+first+second+third+fourth
                    new_gt="gt_"
                    img_width, img_height = image.size   # Get dimensions
                    with open('Art_Dataset/Nobri_New_gt/{}{}.txt'.format(new_gt,str_3),'w',encoding='utf-8') as txtfile:
                        drawer = ImageDraw.Draw(image)

                        flists =[f for f in listdir("OCR-Fonts/") if isfile(join("OCR-Fonts/", f))]
                          # dynamic choice of font 
                        fontlist = np.random.choice(flists,1,replace = False)
                        fsize = np.random.choice(fontsize,2,replace = False)
                        image_size = (img_width,img_height)

                        font = ImageFont.truetype("OCR-Fonts/"+fontlist[0], fsize[0])
                        ascent, descent = font.getmetrics()

                        ###  first word 
                        width1, height1 = drawer.textsize( first, font=font)
                        (width, baseline), (offset_x, offset_y) = font.font.getsize(first)
                        ab = font.getmask(first).getbbox()
                        offset = font.getoffset(first)
                        left_width1, left_height1 =3,1   
                        drawer.text(( left_width1,  left_height1), first, fill=color, font=font)
                        # drawing rectangle surrounding text

                        lengthx1 = left_width1+ab[2] #+ width1 
                        lengthy1 = left_height1 + height1


                        txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width1+ab[0],left_height1+offset[1],
                                                                            lengthx1,left_height1+offset[1],
                                                                            lengthx1,lengthy1,
                                                                            left_width1+ab[0],lengthy1,first
                                                                           ))
                        ##### second word
                        width2, height2 = drawer.textsize( second, font=font)
                        ab = font.getmask(second).getbbox()
                        offset = font.getoffset(second)
                        left_width2, left_height2 =lengthx1-ab[0]+char_gap_1,left_height1
                        drawer.text(( left_width2,  left_height2), second, fill=color, font=font)
                        # drawing rectangle surrounding text

                        lengthx2 = left_width2 +ab[2]
                        lengthy2 = left_height2 + height2
                        txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width2+ab[0],left_height2+offset[1],
                                                                            lengthx2,left_height2+offset[1],
                                                                            lengthx2,lengthy2,
                                                                            left_width2+ab[0],lengthy2,second
                                                                           ))
                        #print('lengthx2:{}'.format(lengthx2))
                        ######  third word
                        width3, height3 = drawer.textsize( third, font=font)
                        ab = font.getmask(third).getbbox()
                        left_width3, left_height3 =lengthx2+char_gap_3-ab[0],left_height2  
                        drawer.text(( left_width3,  left_height3), third, fill=color, font=font)
                        # drawing rectangle surrounding text
                        offset = font.getoffset(third)
                        lengthx3 = left_width3 +ab[2]
                        lengthy3 = left_height3 + height3
                        txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width3+ab[0],left_height3+offset[1],
                                                                            lengthx3,left_height3+offset[1],
                                                                            lengthx3,lengthy3,
                                                                            left_width3+ab[0],lengthy3,third
                                                                           ))

                        ######  fourth word
                        width4, height4 = drawer.textsize( fourth, font=font)
                        ab = font.getmask(fourth).getbbox()
                        left_width4, left_height4 =lengthx3-ab[0]+char_gap_2,left_height3 
                        drawer.text(( left_width4,  left_height4), fourth, fill=color, font=font)
                        # drawing rectangle surrounding text
                        offset = font.getoffset(fourth)
                        lengthx4 = left_width4 + ab[2]#+1
                        lengthy4 = left_height4 + height4
                        txtfile.write("{},{},{},{},{},{},{},{},{}\n".format(left_width4+ab[0],left_height4+offset[1],
                                                                            lengthx4,left_height4+offset[1],
                                                                            lengthx4,lengthy4,
                                                                            left_width4+ab[0],lengthy4,fourth
                                                                           ))

                    image.save('Art_Dataset/Nobri_New_img/{}.png'.format(str_3))
                    img_number+=1
```

```python
from PIL import Image

img_dir = 'Art_Dataset/Nobri_New_img/'
gt_dir = 'Art_Dataset/Nobri_New_gt/'

img_list = [file for file in os.listdir(img_dir) if file.endswith('.png')]
txt_list = [file for file in os.listdir(gt_dir) if file.endswith('.txt')]
print(len(img_list))
print(len(txt_list))
```

```python
num = 1
for i in img_list:
    img = Image.open(img_dir+i)
    txt_name ='gt_'+i[:-4]+'.txt'
    with open(gt_dir+txt_name) as f:
        for line in f.read().splitlines():
            cor = line.split(',')
            x1 = int(cor[0])
            y1 = int(cor[1])
            x2 = int(cor[4])
            y2 = int(cor[5])
            image = img.crop((x1,y1,x2,y2))
            image.save('Art_Dataset/store_test_image/{}.png'.format(num))
            num=num+1
```
